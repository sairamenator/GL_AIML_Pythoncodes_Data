```
%matplotlib inline
# Load/Prep the Data
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# split_condition1 = y_train <5
# split_condition2 = y_train >=5
# labels = y_train[[y_train[5]]
# x_train_Ind04 = x_train [[split_condition1]]
#x_test_Ind04 = x_test [[y_test<5]]
# y_train_Ind04 = y_train[split_condition1]
# y_test_Ind04 = y_test[y_test<5]

# x_train_Ind59 = x_train [split_condition2]
# x_test_Ind59 = x_test [y_test>=5]
# y_train_Ind59 = y_train[split_condition2]
# y_test_Ind59 = y_test[y_test>=5]
```

```
x_train_split0to4 = []
x_test_split0to4 = []
y_train_split0to4 = []
y_test_split0to4 = []

x_train_split5to9 = []
x_test_split5to9 = []
y_train_split5to9 = []
y_test_split5to9= []

for i in np.arange(0,y_train.shape[0]):
  if(y_train[i] <=4):
    x_train_split0to4.append(x_train[i])
    y_train_split0to4.append(y_train[i])
    
  elif(y_train[i]>=5):
    x_train_split5to9.append(x_train[i])
    y_train_split5to9.append(y_train[i])
    
for i in np.arange(0,y_test.shape[0]):
  if(y_test[i] <=4):
    x_test_split0to4.append(x_test[i])
    y_test_split0to4.append(y_test[i])
    
  elif(y_test[i]>=5):
    x_test_split5to9.append(x_test[i])
    y_test_split5to9.append(y_test[i])
    
    
    
    
```

```
x_train_split0to4 = np.array(x_train_split0to4)
x_test_split0to4 = np.array(x_test_split0to4)
y_train_split0to4 = np.array(y_train_split0to4)
y_test_split0to4 = np.array(y_test_split0to4)

print(x_train_split0to4.shape)
print(x_test_split0to4.shape)
print(y_train_split0to4.shape)
print(y_test_split0to4.shape)

x_train_split5to9 = np.array(x_train_split5to9)
x_test_split5to9 = np.array(x_test_split5to9)
y_train_split5to9 = np.array(y_train_split5to9)
y_test_split5to9 = np.array(y_test_split5to9)

print(x_train_split5to9.shape)
print(x_test_split5to9.shape)
print(y_train_split5to9.shape)
print(y_test_split5to9.shape)
```

### 2. Use One-hot encoding to divide y_train and y_test into required no of output classes

```
import pandas as pd
y_train_split0to4 = np.reshape(y_train_split0to4, (25000,))
y_train_split0to4_onehot = pd.get_dummies(y_train_split0to4)
y_test_split0to4 = np.reshape(y_test_split0to4, (5000,))
print(y_train_split0to4_onehot)
y_test_split0to4_onehot = pd.get_dummies(y_test_split0to4)
print(y_test_split0to4_onehot)
```

### 3. Build a sequential neural network model which can classify the classes 0 to 4 of CIFAR10 dataset with at least 80% accuracy on test data

```
from keras.datasets import cifar10, mnist
from keras.models import Sequential# sequential is one type of model; there are graph models as well
from keras.layers import Dense, Activation, Dropout, Flatten, Reshape # Dense is fully connected layer
from keras.layers import Convolution2D, MaxPooling2D #
from keras.utils import np_utils
import pickle # serialization fancy word for storing on disk
from matplotlib import pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (15, 8)
```

```
TRAIN = False
BATCH_SIZE = 32
EPOCHS = 10 

# Define the Type of Model
model2 = Sequential()
# 1st Conv Layer
model2.add(Convolution2D(32, 3, 3, input_shape=(32,32,3)))
model2.add(Activation('relu'))

# 2nd Conv Layer
model2.add(Convolution2D(32, 3, 3))# input shape is known in the previous layer. so we don't give input shape
model2.add(Activation('relu'))

# Fully Connected Layer
model2.add(Flatten())
model2.add(Dense(128))
model2.add(Activation('relu'))


# Layer 1
model2.add(Dense(output_dim=5, bias=True))
model2.add(Activation("softmax"))

# Loss and Optimizer
model2.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Store Training Results
early_stopping = keras.callbacks.EarlyStopping(monitor='val_acc', patience=5, verbose=1, mode='auto')
callback_list = [early_stopping]# [stats, early_stopping]

# Train the model
model2.fit(x_train_split0to4, y_train_split0to4_onehot, nb_epoch=EPOCHS, batch_size=BATCH_SIZE,
        validation_data=(x_test_split0to4, y_test_split0to4_onehot), callbacks=callback_list, verbose=True)
```

```
model2.evaluate(x_test_split0to4, y_test_split0to4_onehot)
```

### 4. In the model which was built above (for classification of classes 0-4 in CIFAR10), make only the dense layers to be trainable and conv layers to be non-trainable

```
#Freezing layers in the model which don't have 'dense' in their name
for layer in model2.layers:
  if('dense' not in layer.name): #prefix detection to freeze layers which does not have dense
    #Freezing a layer
    layer.trainable = False # freezing the layer

#Module to print colourful statements
from termcolor import colored

#Check which layers have been frozen 
for layer in model2.layers:
  print (colored(layer.name, 'blue'))
  print (colored(layer.trainable, 'red'))
```
